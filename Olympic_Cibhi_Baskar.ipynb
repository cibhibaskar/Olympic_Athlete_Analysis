### Don't forget to drop duplicates first

### What is the number of winners with the gold medal in each city?

```python
#Filter the rows with Gold medal
df_gold = df[df['Medal'] == 'Gold']
```

```python
#Get the count of rows for each city
df_gold = df_gold.groupby(['City', 'Medal']).size()
df_gold = pd.DataFrame(df_gold)
df_gold.head()
```

```python
df_gold.columns = ['Count']
```

```python
df_gold = df_gold.reset_index().sort_values(by=['Count'], ascending=False)
df_gold.head()
```

```python
df_gold.columns
```

```python
df_gold.drop(['Medal'], axis = 1, inplace = True)
```

```python
#Plot it in the bar plot
width = 120
height = 50
plt.figure(figsize=(width, height))

plot = plt.bar(df_gold['City'], df_gold['Count'], color='orange')

plt.bar_label(plot, label_type='center', fontsize=50)

plt.xlabel('City', fontsize=120)
plt.ylabel('Gold Count', fontsize=120)
plt.title('Gold Count by City', fontsize=120, fontweight='bold')
plt.xticks(rotation=90, size=90)
plt.yticks(size=120)

plt.show()
```

### Has the number of athletes changed over time for Summer and Winter?

```python
#Filter the summer season
df_summer = df[df['Season']=='Summer']
df_summer = df_summer[['Year', 'ID', 'Season']]
df_summer.head()
```

```python
#Filter the Winter Season
df_winter = df[df['Season']=='Winter']
df_winter = df_winter[['Year', 'ID', 'Season']]
df_winter.head()
```

```python
#Get unique rows to get the number of athlets each year for summer
df_summer = df_summer.groupby(['Year']).nunique()
df_summer.head()
```

```python
#Get unique rows to get the number of athlets each year for winter
df_winter = df_winter.groupby(['Year']).nunique()
df_winter.head()
```

```python
df_summer.drop(['Season'], axis = 1, inplace = True)
df_summer.columns = ['Count']
df_summer = df_summer.reset_index().sort_values(by=['Year'], ascending=True)
df_summer.head()
```

```python
df_winter.drop(['Season'], axis = 1, inplace = True)
df_winter.columns = ['Count']
df_winter = df_winter.reset_index().sort_values(by=['Year'], ascending=True)
df_winter.head()
```

```python
#Plot a subplot plotting both summer and winter number of Athlets
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 10))

# Plotting on ax1 (Summer data)
ax1.plot(df_summer['Year'], df_summer['Count'], marker='o', color='r')
ax1.set_xlabel('Year')
ax1.set_ylabel('Count')
ax1.set_title('Number of Athlets Over the Years (Summer)')

# Plotting on ax2 (Winter data)
ax2.plot(df_winter['Year'], df_winter['Count'], marker='o', color='b')
ax2.set_xlabel('Year')
ax2.set_ylabel('Count')
ax2.set_title('Number of Athlets Over the Years (Winter)')
```
